## EDA- Bivariate and Multivariate

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
flights= sns.load_dataset('flights')
flights.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

## Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'])
```

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'],hue=df['day'],size=df['size'],style=df['smoker'])
```

## BarPlot (Numerical -Categorical)

```python
sns.barplot(x=df['time'],y=df['total_bill'])
```

## Boxplot (Numerical -Categorical)

```python
sns.boxplot(x=flights['month'],y=flights['passengers'])
```
